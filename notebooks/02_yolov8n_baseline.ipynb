```python
import kagglehub

path = kagglehub.dataset_download("aliabdelmenam/rdd-2022")

print(path)
```

```python
yaml_content = """
path: /kaggle/input/datasets/aliabdelmenam/rdd-2022/RDD_SPLIT

train: train/images
val: val/images
test: test/images

names:
  0: longitudinal crack
  1: transverse crack
  2: alligator crack
  3: other corruption
  4: pothole
"""

with open("data.yaml", "w") as f:
    f.write(yaml_content)

print("data.yaml created")
```

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")
```

```python
results = model.train(
    data="data.yaml",
    epochs=20,
    imgsz=640,
    batch=16,
    save_period=1,
    workers=4,
    project="/kaggle/working",
    name="yolov8n_baseline"
)
```

## Results Summary

The YOLOv8n baseline model was trained for 20 epochs on the RDD2022 dataset.

Final Results:

- Precision: 0.579
- Recall: 0.516
- mAP@0.5: 0.539
- mAP@0.5:0.95: 0.285

Observations:

- The model achieved stable convergence without significant overfitting.
- Other Corruption achieved the highest class-wise performance.
- Pothole detection remained the most challenging category.
- Small object detection continues to be a limitation, motivating the use of SAHI in subsequent experiments.
